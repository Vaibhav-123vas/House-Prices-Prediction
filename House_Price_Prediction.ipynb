                                                House Pricing Prediction  

  Explanation:
Load the dataset: Reads the CSV file into a pandas DataFrame.
Select relevant features: Chooses the square footage (GrLivArea), number of bedrooms (BedroomAbvGr), and number of bathrooms (FullBath).
Handle missing values: Fills any missing values in the features with the mean of the column.
Split the data: Divides the dataset into training (80%) and testing (20%) sets.
Train the model: Uses the training data to fit a linear regression model.
Make predictions: Predicts house prices on the test data.
Evaluate the model: Calculates and prints the Mean Squared Error and R^2 Score.
Plotting: Creates a scatter plot to visualize the relationship between actual :and predicted prices.
Predicting house prices with sample data: Predicts house prices for given sample data (e.g., houses with specific square footage, bedrooms, and bathrooms).                                              



```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
 #Load the dataset
data = pd.read_csv('train[1].csv')
```

```python
#Handling relevant features
features = data[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
target = data['SalePrice']
```

```python
#Handling Missing values
features.fillna(features.mean(), inplace=True)
```

```python
#Splitting into training and testing data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
#Fitting the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
 #Plotting the results
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
#Predicting house prices with given features
sample_data = pd.DataFrame({
    'GrLivArea': [1500, 2500],
    'BedroomAbvGr': [3, 4],
    'FullBath': [2, 3]
})

sample_predictions = model.predict(sample_data)
print("Predicted Prices for sample data:", sample_predictions)
```
